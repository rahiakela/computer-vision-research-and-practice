```python
from diffusers import DiffusionPipeline
from diffusers import EulerDiscreteScheduler
```

##DiffusionPipeline

```python
# Load the model
pipeline = DiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", use_safetensors=True)
```

```python
pipeline
```

```python
# Send model to GPU
pipeline.to("cuda")
```

```python
# Now you can pass a text prompt to the pipeline to generate an image
image = pipeline("An image of a squirrel in Picasso style").images[0]
```

```python
image
```

```python
image = pipeline("An image of an Indian warrior in modern style").images[0]
```

```python
image
```

##Swapping schedulers

```python
pipeline = DiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", use_safetensors=True)
pipeline.scheduler = EulerDiscreteScheduler.from_config(pipeline.scheduler.config)
```

```python
pipeline.scheduler
```

```python
pipeline.to("cuda")

image = pipeline("An image of a squirrel in Picasso style").images[0]
```

```python
image
```

```python
image = pipeline("An image of an Indian warrior in modern style").images[0]
```

```python
image
```
